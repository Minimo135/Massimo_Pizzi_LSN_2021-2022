# Esercitazione 11

Con la presente esercitazione vogliamo allenare una rete neurale (NN) a eseguire _fit_ di dati. Poiché si avrà necessità di stimare le prestazioni della rete, i dati vengono generati a partire dalla soluzione al problema, cioè dalla funzione che meglio li interpola.

# Esercizio 11.1

Iniziamo dalla più semplice delle funzioni, cioè la retta. Prendiamo per questo esercizio $g(x) = 2x + 1$. Dati dei punti distribuiti unuiformemente attorno a questa funzione, la rete proverà a effettuare una regressione e trovare quella funzione che _meglio_ approssima questa distribuzione. Naturalmente tale funzione è proprio $g(x)$, perciò, confrontandola con quella dedotta dalla rete, avremo modo di valutare quantitativamente la bontà della regressione.

In primo luogo è necessario _addestrare_ la rete affinché "impari" la relazione che implicitamente governa la distribuzione dei dati. A tale scopo vengono generati $N$ punti con ascisse $x_{train, i}$ distribuite uniformemente nell'intervallo $(-1,1)$ e ordinate $y_{train,i} = g(x_{train, i}) + \xi_i$ , dove $\xi_i$ è una variabile aleatoria distribuita normalmente con varianza $\sigma$. La rete neurale, prese in ingresso le stesse variabili indipendenti $x_{train, i}$, produrrà altrettante $f(x_{train,i})$. La forma funzionale di $f$ dipende dall'architettura della rete (dal numero di neuroni artificiali e di strati e dalle connessioni tra neuroni), ma per ora costruiamo una rete costituita da un singolo neurone. In questo caso, $f$ è semplice e anche adatta al problema, infatti è la funzione lineare

$$ f_{w,b}(x) = wx+b.$$

Dunque il neurone di fatto è una _funzione_. Perché $f$ sia proprio la funzione che stiamo cercando, la distanza dei valori prodotti dalla rete ($f_{w,b}(x_{train,i})$) dai quelli dati in ingresso ($y_{train,i}$) deve essere minima. Valutiamo questa discrepanza usando lo scarto quadratico medio

$$ S(w,b) = \frac{1}{N}\sum_{i = 1}^{N}( y_{train,i} - f_{w,b}(x_{train,i}) )^2 .$$

La rete, partendo da dei valori iniziali per $w$ e $b$, proverà a cercare quelli per qui la funzione scritta sopra sia minima. Nel gergo dell'apprendimento automatico, la funzione che deve essere ottimizzata in fase di addestramento (o _training_) della rete è detta _loss function_ o funzione di costo. Sono diversi gli algoritmi di ricerca del minimo, ma in questa esercitazione useremo lo _Stochastic Gradient Descend_ (SGD), che è già implementato da Keras. I dati in ingresso vengono casualmente raggruppati in _batch_ di dimensione minore e ordinatamente dati in input alla rete. Questa cerca dei nuovi valori per $w$ e $b$ muovendosi in direzione opposta al gradiente di $S(w,b)$ di un certo passo. Questa operazione viene eseguita per ogni batch e, terminati gli input, un'_epoca_ dell'addestramento si conclude.

Al termine di ogni epoca verrà salvato il valore attuale della _loss function_, che è indicativo solo di quanto bene la rete abbia adattato i propri parametri ai dati con cui si è allenata. Perciò se intendiamo valutare le sue effettive prestazioni, è utile calcolare la _loss function_ usando anche dati che la rete non abbia mai "visto". A tale scopo viene generato un secondo set di $M$ punti con ascisse $x_{valid, i}$ e ordinate $y_{valid, i} = g(x_{valid, i}) + \xi_i$ e ad ogni epoca lo scarto quadratico medio è calcolato anche in rapporto a questi. Di solito si indica questo valore con _validation loss function_. Ci riferiamo agli $M$ dati con cui la rete verifica le proprie prestazioni con _dati di test_ (o _validation data_) e ne prendiamo - in numero - meno dei dati di training, in particolare $M = \text{ceil}(0.1\times N) + 5$

Si valuterà il rendimento della rete per diversi valori di $N$ (10, 100, 1000, 10000) e $\sigma$ (0.0, 0.1, 0.25, 0.5)

```python
# target parameters of f(x) = m*x + b
m = 2 # slope
b = 1 # intersect
```

```python
import numpy as np

# generate training inputs
np.random.seed(0)
N = [10**i for i in range(1, 5)]; print(N)
x_train = [ np.random.uniform(-1, 1, N[i]) for i in range(len(N))] #Training inputs of (10, 100, 1000) points
x_valid = [np.random.uniform(-1, 1, np.ceil(n*0.1).astype(int) + 5) for n in N] #Validation data, 10% on x_train points
x_target = np.random.uniform(-1, 1, 50)
x_target.sort()
y_target = m * x_target + b # ideal (target) linear function

sigma = np.array([0, 0.1, 0.25, 0.5]) # noise standard deviation
y_train = [ [np.random.normal(m * xtr + b, s) for xtr in x_train] for s in sigma] # actual measures from which we want to guess regression parameters
y_valid = [ [np.random.normal(m * x_valid + b, s) for x_valid in x_valid ] for s in sigma]
```

```python
from numpy.lib.polynomial import poly
# plot validation and target dataset
import matplotlib.pyplot as plt

fig, plts = plt.subplots(sigma.size,1)
for i in range(sigma.size):
    for j in range(len(x_train)-1):
        plts[i].scatter(x_train[j], y_train[i][j], marker = '.',  label='training data ($x_{train,i}, y_{train,i}$), N = ' + str(x_train[j].size))
        plts[i].scatter(x_valid[j], y_valid[i][j], label='validation data $(x_{valid,i},y_{valid,i})$')
    plts[i].legend()
    plts[i].grid(True)
    plts[i].set_title('Sigma = ' + str(sigma[i]))

fig.set_figwidth(15)
fig.set_figheight(5*sigma.size)
```

Costruiamo tanti modelli di rete, quante sono le combinazioni di $N$ e $\sigma$ che vogliamo studiare, perciò in totale alleneremo sedici modelli. Aumentiamo il parametro della <code>learning_rate</code> dello SGD rispetto al valore predefinito, per ottenere una ricerca più rapida del punto di minimo. Per essere sicuri di aver costruito i modelli correttamente, ne visualizziamo la struttura con il metodo <code>summary()</code>.

#### ------------------------------------------------------------------------------------------------------------------------------------------------

```python
# compose the NN model
import tensorflow as tf
from tensorflow import keras

from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense, Activation
from tensorflow.keras import backend as K
from tensorflow.keras.utils import get_custom_objects

model = [ [tf.keras.Sequential() for j in range(len(x_train))] for i in range(sigma.size) ]
for i in range(sigma.size):
    for j in range(len(x_train)):
        #Aggiungiamo un singolo neurone
        model[i][j].add(Dense(1, input_shape=(1,)))
        # Utilizziamo un algoritmo di gradient descend per ottimizzare la loss function
        model[i][j].compile(optimizer=tf.keras.optimizers.SGD(learning_rate = 0.02), loss='mse', metrics=['mse'])
```

```python
# get a summary of our composed model
model[0][0].summary()
```

Come atteso, sono solo due i parametri da ottimizzare, ossia $w$ e $b$. Ora alleniamo ciascun modello sui rispettivi dati per 30 epoche con _batch_ di dimensione diversa in base al numero di dati in ingresso, di modo che ad ogni epoca il numero di batch è sempre lo stesso e sia $10$.

```python
# fit the model using training dataset
# over 10 epochs of 32 batch size each
# report training progress against validation data
batches = [1, 10, 100, 1000, 10000]
history = [ [model[i][j].fit(x=x_train[j], y=y_train[i][j],
                             batch_size=batches[j],
                             epochs=30,
                             shuffle=True,
                             validation_data=(x_valid[j], y_valid[i][j]) ) for j in range(len(x_train))] for i in range(sigma.size) ]
```

#### ---------- Eseguire la cella sotto per salvare i modelli e la storia ---------

```python
save_model_path= [[('./Saved/models/mdl_sigma' + str(sigma[i]) + '_N' + str(N[j]) + '.h5') for j in range(len(N))] for i in range(sigma.size)]
#print(save_model_path)
for i in range(sigma.size):
    for j in range(len(N)):
        model[i][j].save(filepath=save_model_path[i][j], include_optimizer=True)
        
#Salvo la storia...
save_hist_path = [[('./Saved/histories/hist_sigma' + str(sigma[i]) + '_N' + str(N[j]) + '.dat') for j in range(len(N))] for i in range(sigma.size)]
for i in range(sigma.size):
    for j in range(len(N)):
        history_ij = np.c_[history[i][j].history['loss'], history[i][j].history['val_loss']]
        np.savetxt(save_hist_path[i][j], history_ij)
        
        
# history[i][j].history['loss']
# history[i][j].history['val_loss']
```

Terminato l'addestramento, mostriamo i valori finali dei parametri del neurone artificiale che la rete ha trovato nei diversi casi

```python
# return weights and biases
for i in range(sigma.size):
    print('Sigma = '  + str(sigma[i]))
    for j in range(len(x_train)):
        print('  N = ' + str(10**(j+1)))
        print('   w = ' +  str(model[i][j].get_weights()[0]) + ' - atteso: ' + str(2))
        print('   b = ' +  str(model[i][j].get_weights()[1]) + ' - atteso: ' + str(1))
```

Keras dispone anche di un metodo <code>evaluate()</code>, che permette di valutare la _loss function_ su un determinato insieme di dati, in questo caso i punti $(x_{valid,i}, y_{valid,i})$, ossia gli stessi usati in fase di addestramento per valutare la _validation loss function_.

```python
# evaluate model
batches = [1, 10, 100, 1000, 10000]
score = [ [ model[i][j].evaluate(x_valid[j], y_valid[i][j], batch_size=batches[j], verbose=1) for j in range(len(x_train)) ] for i in range(sigma.size) ]

# print performance
print()
for i in range(sigma.size):
  print('Sigma = ' + str(sigma[i]))
  for j in range(len(x_train)):
    print('  N = ' + str(x_train[j].size))
    print('    Test loss:', score[i][j][0])
```

Sarebbe interessante valutare il rendimento della rete, confrontandone le previsioni con quei punti che giacciono esattamente sulla retta $y = 2x+1$. Eseguiamo allora ancora una volta <code>evaluate()</code> su 50 punti $(x_{target,i}, 2x_{target,i}+1)$, dove $x_{target,i}$ sono distribuiti uniformemente tra -1 e 1.

```python
# evaluate model with the exact curve
score_tar = [[model[i][j].evaluate(x_target, y_target, batch_size=batches[j], verbose=1) for j in range(len(x_train)) ] for i in range(sigma.size)]

# print performance
print()
for i in range(sigma.size):
  print('Sigma = ' + str(sigma[i]))
  for j in range(len(x_train)):
    print('  N = ' + str(x_train[j].size))
    print('    Test loss:', score_tar[i][j][0])
```

Per una lettura più chiara di questi risultati, costruiamo due serie di grafici che mostrino la variazione della _loss function_ a $N$ costante e a $\sigma$ costante, per il confronto sia coi dati di test, sia con quelli esatti.

```python
fig, plts = plt.subplots(1, sigma.size)
for i in range(sigma.size):
    plts[i].plot(np.log10(N), [score[i][j][0] for j in range(len(x_train))], 'r.:', label = 'Validation data')
    plts[i].plot(np.log10(N), [score_tar[i][j][0] for j in range(len(x_train))], 'b.:', label = 'Exact data')
    plts[i].legend(loc='best')
    plts[i].grid(True)
    plts[i].set_ylabel('Loss')
    plts[i].set_xlabel('log10(N)')
    plts[i].set_title('Sigma = ' + str(sigma[i]))

fig.set_figwidth(17)
fig.set_figheight(5)

fig2, plts2 = plt.subplots(1, len(N))
for i in range(len(N)):
    plts2[i].plot(sigma, [score[j][i][0] for j in range(sigma.size)], 'r.:', label = 'Validation data')
    plts2[i].plot(sigma, [score_tar[j][i][0] for j in range(sigma.size)], 'b.:', label = 'Exact data')
    plts2[i].legend(loc='best')
    plts2[i].grid(True)
    plts2[i].set_ylabel('Loss')
    plts2[i].set_xlabel('sigma')
    plts2[i].set_title('N = ' + str(10**(i+1)))

fig2.set_figwidth(17)
fig2.set_figheight(5)
```

Quando valutata rispetto ai dati di test, osserviamo una chiara tendenza della rete a diventare meno precisa all'aumentare di $\sigma$, come ragionevolmente ci si potrebbe aspettare. I dati infatti deviano maggiormente dalla distribuzione lineare rispetto a cui sono stati casualmente perturbati e perciò lo scarto quadratico medio non potrà che aumentare, per quanto si tenti di ottimizzarlo. Sempre in rapporto ai dati di test, la rete non sembra migliorare nel trovare la funzione interpolatrice al variare dei dati in ingresso. Anche questo non deve sorprendere, poiché lo scarto dipende dalla varianza e non dal numero di dati in ingresso.

Per quanto riguarda il rendimento in rapporto ai dati "esatti" che giacciono sulla retta di soluzione, maggiore è il numero di dati in gresso, minore è lo scarto quadratico medio. Questo significa che la rete, nonostante faccia fatica ad ottimizzare la loss function per i dati perturbati, ha "capito" quale sia la funzione interpolante.

Infine, per avere una panoramica complessiva, mostriamo i grafici delle valutazioni della _loss function_ in funzione delle epoche dell'addestramento, sia sui dati di training, sia su quelli di test. Infine mostriamo i dati interpolati dlla rete partendo dai valori $x_{valid,i}$

## Esercizio 11.1

Con questo esercizio si vuole estendere la rete costruita perché impari ad eseguire il _fit_ di dati distribuiti attorno al seguente polinomio di terzo grado:
$$
g(x)=4-3x-2x^2+3x^3
$$
per $x \in [-1,1]$.

Si vogliono sfruttare le potenzialità della rete, aggiungendo neuroni e strati di neuroni. Verranno provati diversi algoritmi per l'ottimizzazione della funzione di costo e anche diverse funzioni costo.

Prima di ideare la nuova architettura della rete neurale, introduciamo le funzioni di attivazione. Queste agiscono sull'output di un neurone in maniera non lineare. Senza il loro intervento, una rete non sarà mai in grado di effettuare regressioni di funzioni generiche! Ogni neurone, come abbiamo visto, ha infatti un comportamento intrinsecamente lineare. Keras fornisce diverse funzioni di attivazione di default: <code>relu</code>, <code>elu</code>, <code>sigmoid</code>, <code>tanh</code>, ecc... Ma è anche possibile inventare una funzione personalizzata.

Sulla falsa riga dell'esercizio precedente, alleneremno la rete su $N$ dati di training e la valuteremo sia ad ogni epoca, che alla fine dell'addestramento su $M$ dati di test. Verificheremo inoltre la discrepanza dalla soluzione esatta attesa del _fit_, cioè la discrepanza dal polinomio $g(x)$

```python
# generate training inputs
np.random.seed(0)
x_train = np.random.uniform(-1, 1, 10000) #Training inputs of (10, 100, 1000) points
x_valid = np.random.uniform(-1, 1, 100) #Validation data, 10% on x_train points
x_valid.sort()
y_target = 4 - 3*x_valid - 2*x_valid**2 + 3*x_valid**3 # ideal (target) linear function

sigma = 0.1 # noise standard deviation
y_train = np.random.normal(4 - 3*x_train - 2*x_train**2 + 3*x_train**3, sigma) # actual measures from which we want to guess regression parameters
y_valid = np.random.normal(4 - 3*x_valid - 2*x_valid**2 + 3*x_valid**3, sigma)
```

```python
from numpy.lib.polynomial import poly
# plot validation and target dataset
import matplotlib.pyplot as plt

plt.scatter(x_train, y_train, marker = '.',  label='training data ($x_{train,i}, y_{train,i}$), N = 10000')
plt.scatter(x_valid, y_valid, label='validation data $(x_{valid,i},y_{valid,i})$')
plt.xlabel('x')
plt.xlabel('y')
plt.legend()
plt.grid(True)
plt.title('Sigma = ' + str(sigma))

fig.set_figwidth(15)
fig.set_figheight(5*1)
```

Sperimentiamo diverse funzioni di attivazione, che nel caso di un modello di rete sequenziale agiscono sui neuroni di un intero strato. L'output deve essere necessariamente un singolo neurone.

Aggiungere strati equivale ad effettuare composizioni delle funzioni di attivazione, mentre ampliare uno stesso strato significa combinare linearmente la stessa funzione di attivazione. Iniziamo con un singolo strato di due neuroni. Si ricorda inoltre che, prima che venga applicata la funzione di attivazione, l'input è trasformato linearmente secondo due parametri $w$ e $b$.

Proveremo due funzioni di attivazione dotate di _plateau_ (tangente iperbolica e sigmoide) e due senza (relu ed elu). La funzione di costo è sempre lo scarto quadratico medio.

In seguito sono mostrati i grafici della variazione della _loss function_ per le varie funzioni di attivazioni e il grafico dei punti estrapolati dalla rete una volta addestrata (con il confronto con la funzione target).

```python
import tensorflow as tf
from tensorflow import keras

from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense, Activation, Dropout
from tensorflow.keras import backend as K
from tensorflow.keras.utils import get_custom_objects

acfun = ['tanh', 'sigmoid', 'relu', 'elu']

model = [tf.keras.Sequential(
    [
        Dense(2, activation = fun, input_shape=(1,) ),
        Dense(1, activation = None ),
    ]
) for fun in acfun]

for m in model:
    m.compile(optimizer=tf.keras.optimizers.SGD(learning_rate = 0.01), loss='mse', metrics=['mse'])
    
model[0].summary()
```

```python
history2 = [ m.fit(x=x_train, y=y_train,
                             batch_size=16,
                             epochs=20,
                             shuffle=True,
                             validation_data=(x_valid, y_valid) ) for m in model]
```

```python
fig, plts = plt.subplots(1, len(acfun))
for j in range(len(acfun)):
    plts[j].plot(history2[j].history['loss'])
    plts[j].plot(history2[j].history['val_loss'])
    plts[j].legend(['Loss function - train', 'Loss function - test'], loc='best')
    plts[j].grid(True)
    plts[j].set_ylabel('Loss')
    plts[j].set_xlabel('Epoch')
    plts[j].set_title('Activation function: ' + str(acfun[j]))

fig.set_figwidth(15)
fig.set_figheight(5)
```

Questi primi risultati suggeriscono quale possa essere la funzione di attivazione più adatta per fare regressioni di funzioni polinomiali, usando solo pochi neuroni, ossia la tangente iperbolica.

```python
scores2 = [m.evaluate(x_valid, y_valid, batch_size=16, verbose=1) for m in model]
for i in range(len(acfun)):
    print('Activation function: ' + str(acfun[i]))
    print(' Test loss = ' + str(scores2[i][0]) )
fig, plts = plt.subplots(1, len(acfun))
for i in range(len(acfun)):
    x_predicted = np.random.uniform(-1, 1, 100)
    y_predicted = model[i].predict(x_predicted).reshape(100)
    print(x_predicted.shape); print(y_predicted.shape);
    plts[i].scatter(x_predicted, y_predicted, color='r')
    plts[i].plot(x_valid, y_target)
    plts[i].grid(True)
    plts[i].set_title('Activation function: ' + str(acfun[i]))
    plts[i].set_xlabel('x')
    plts[i].set_ylabel('y')
    
fig.set_figwidth(15)
```

La funzione _relu_ è lineare per $x$ positive, nulla per $x$ negative, perciò è ragionevole aspettarsi che una combinazione di due sole funzioni di questo tipo non è sufficiente ad approssimare un polinomio di terzo grado. Lo stesso discorso si applica per la funzione _elu_, una versione alternativa che per $x$ negative si comporta come un esponenziale, tuttavia, presentanto un raccordo liscio, si presta meglio ad essere adattata per interpolare funzioni con concavità.

Proviamo nuovamente decuplicando il numero di neuroni nello strato

```python
model = [tf.keras.Sequential(
    [
        Dense(20, activation = fun, input_shape=(1,) ),
        Dense(1, activation = None ),
    ]
) for fun in acfun]

for m in model:
    m.compile(optimizer=tf.keras.optimizers.SGD(learning_rate = 0.01), loss='mse', metrics=['mse'])
model[0].summary()
history3 = [ m.fit(x=x_train, y=y_train,
                             batch_size=16,
                             epochs=20,
                             shuffle=True,
                             validation_data=(x_valid, y_valid) ) for m in model]
```

```python
fig, plts = plt.subplots(1, len(acfun))
for j in range(len(acfun)):
    plts[j].plot(history3[j].history['loss'])
    plts[j].plot(history3[j].history['val_loss'])
    plts[j].legend(['Loss function - train', 'Loss function - test'], loc='best')
    plts[j].grid(True)
    plts[j].set_ylabel('Loss')
    plts[j].set_xlabel('Epoch')
    plts[j].set_title('Activation function: ' + str(acfun[j]))

fig.set_figwidth(15)
fig.set_figheight(5)
```

```python
scores3 = [m.evaluate(x_valid, y_valid, batch_size=16, verbose=1) for m in model]
for i in range(len(acfun)):
    print('Activation function: ' + str(acfun[i]))
    print(' Test loss = ' + str(scores3[i][0]) )

fig, plts = plt.subplots(1, len(acfun))
for i in range(len(acfun)):
    x_predicted = np.random.uniform(-1, 1, 100)
    y_predicted = model[i].predict(x_predicted).reshape(100)
    print(x_predicted.shape); print(y_predicted.shape);
    plts[i].scatter(x_predicted, y_predicted, color='r')
    plts[i].plot(x_valid, y_target)
    plts[i].grid(True)
    plts[i].set_title('Activation function: ' + str(acfun[i]))
    plts[i].set_xlabel('x')
    plts[i].set_ylabel('y')
    
fig.set_figwidth(15)
```

La tangente iperbolica continua ad essere la funzione di attivazione che meglio permetta alla rete di ottimizzare la funzione di costo.

Spesso in fisica è noto che certi dati seguano una certa legge. Se supponiamo che stiamo cercando un polinomio di terzo grado, possiamo usare come funzione di attivazione proprio $x^3$. Bastano allora due soli neuroni perché si possa rappresentare qualunque polinomio di terzo grado. Infatti questi dipendono da $7$ parametri e per determinare un polinomio di terzo grado ne basta fissare solo $4$. Aggiungere altri neuroni allo strato può migliorare il tempo che impiega la rete ad otimizzare la funzione, perché diverse combinazioni dei parametri porterebbero alla stessa soluzione.

Dopo una serie di tentativi, si è scoperto che l'addestramento non riesce a trovare i parametri  ottimali qualora si usi l'ottimizzatore SGD con le impostazioni predefinite, poiché troppo spesso si ferma su minimi locali. Due le soluzioni trovate: ridurre il numero di input per batch, affinché la ricerca dei parametri si muova maggiormente nello spazio dei parametri, assieme ad un aumento della _learning rate_; in alternativa, cambiare l'ottimizzatore e usare _Adadelta_ risolve il problema in tempo minore. _Adadelta_ aggiorna la _learning rate_ durante la ricerca, in base al valore del gradiente attuale e di quelli passati, partendo da un valore iniziale impostato a $0.5$.

```python
# compose the NN model
import tensorflow as tf
from tensorflow import keras

from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense, Activation, Dropout
from tensorflow.keras import backend as K
from tensorflow.keras.utils import get_custom_objects

pow3 = lambda x: tf.pow(x,3)

modelpow = tf.keras.Sequential()
#Aggiungiamo uno strato di 200 neuroni
modelpow.add(Dense(3, activation = pow3, input_shape=(1,) ))
modelpow.add(Dense(1, activation = None ))
# model.add(Dense(1))
# Utilizziamo un algoritmo di gradient descend per ottimizzare la loss function
modelpow.compile(optimizer=tf.keras.optimizers.Adadelta(learning_rate = 0.5), loss='mse', metrics=['mse'])
# get a summary of our composed model
modelpow.summary()
```

```python
history2 = modelpow.fit(x=x_train, y=y_train,
                             batch_size=32,
                             epochs=20,
                             shuffle=True,
                             validation_data=(x_valid, y_valid) )
```

```python
fig, plts = plt.subplots(1, 2)
for j in range(2):
    plts[j].plot(history2.history['loss'])
    plts[j].plot(history2.history['val_loss'])
    plts[j].legend(['Loss function - train', 'Loss function - test'], loc='best')
    plts[j].grid(True)
    plts[j].set_ylabel('Loss')
    plts[j].set_xlabel('Epoch')
    plts[j].set_title('Sigma = ' + str(sigma) + ', N = ' + str(10000) )

fig.set_figwidth(15)
fig.set_figheight(5)
```

## Approfondimento

Nel secondo esercizio, è stata costruita una rete che effettua regresioni usando in pratica combinazioni di funzioni sinusoidali. Questo approccio suggerisce che si possa costruire una rete che provi ad approssimare una distribuzione di dati usando funzioni polinomiali (esiste sempre un polinomio che approssima con precisione arbitraria una funzione cintinua su un intervallo limitato). Per costruirla, il modello sequenziale di rete non basta. Viene allora sperimentato l'utilizzo del _Functional API_, un modo per creare modelli che non siano sequenziali. L'idea per la rete è la sequente: il valore in input viene passato a $K$ livelli "paralleli", ciascuno dei quali con una funzione potenza come attivazione: $x\mapsto x^n$, con $n = 0, 1, 2, \dots, K$. Non si tratta di un unico strato denso, poiché Keras non permette di personalizzare la funzione di attivazione _per neurone_. Infine l'output della rete raccoglie le uscite dai $K$ strati.

```python
#l'input è un singolo valore
inputs = keras.Input(shape=(1,))
K = 4
#costruisco + 1 strati con funzione di attivazione x^n
pol_0 = Dense(1)(inputs)
pol_n = [ Dense(1, activation = lambda x: tf.pow(x, n), use_bias = True)(inputs) for n in range(1, K+1)]
pol_n.append(pol_0)
concatenate = keras.layers.Concatenate()(pol_n)
outputs = Dense(1)(concatenate)

modelPOL = keras.Model(inputs=inputs, outputs=outputs, name = 'polynomial_model')
modelPOL.summary()
modelPOL.compile(optimizer=tf.keras.optimizers.Adam(), loss='mse', metrics=['mse'])
```

```python
historyPOL = modelPOL.fit(x=x_train, y=y_train,
                             batch_size=32,
                             epochs=20,
                             shuffle=True,
                             validation_data=(x_valid, y_valid) )
```

```python
fig, plts = plt.subplots(1, 2)
for j in range(2):
    plts[j].plot(historyPOL.history['loss'])
    plts[j].plot(historyPOL.history['val_loss'])
    plts[j].legend(['Loss function - train', 'Loss function - test'], loc='best')
    plts[j].grid(True)
    plts[j].set_ylabel('Loss')
    plts[j].set_xlabel('Epoch')
    plts[j].set_title('Sigma = ' + str(sigma) + ', N = ' + str(10000) )

fig.set_figwidth(15)
fig.set_figheight(5)

modelPOL.get_weights()
```

```python
x_predicted = np.random.uniform(-3, 3, 100)
x_predicted.sort()
y_predicted = modelPOL.predict(x_predicted).reshape(100)
y_target = 4 - 3*x_predicted - 2*np.power(x_predicted,2) + 3*np.power(x_predicted,3)
print(x_predicted.shape); print(y_predicted.shape);
plt.scatter(x_predicted, y_predicted, color='r')
plt.plot(x_predicted, y_target)
plt.grid(True)
plt.title('Sigma = ' + str(sigma) + ', N = ' + str(10000 ))
plt.xlabel('x')
plt.ylabel('y')
```

Poiché la funzione target è un polinomio di terzo grado, conviene porre $K=3$. Se non è nota la funzione per cui si esegue il fit, un numero di $K$ non oltre $10$ dovrebbe essere sufficiente per approssimare un grande numero di funzioni. In questo caso, tuttavia, la funzione di fit potrebbe non coincidere con quella attesa oltre il dominio di addestramento.

# Esercitazine 9

In questa esercitazione un algoritmo genetico viene impiegato per vercare una soluzione al problema del commesso viaggiatore. Dato un insieme di punti, si vuole trovare quel percorso che passando per tutti ritorna al punto di partenza e abbia lunghezza minore.

Inizieremo con il caso in cui i punti giacciono su una circonferenza, per cui la soluzione è nota e consiste nel percorso che tocca i punti consecutivamente. In seguito lo stesso algoritmo verrà usato nel caso più generale in cui i punti sono dsiposti casualmente all'interno di un quadrato.

L'idea alla base degli algoritmi genetici (o _GA_) nasce dall'osservazione dei processi di selezione naturale, attraverso cui una specie evolve nel tempo: all'interno di una popolazione, gli individui che per ragioni genetiche sono adatti a vivere nell'ambiente, si riproducono - con la tendenza a scegliersi tra loro - e generano una prole che mantenga quelle qualità che rendono la specie adatta a sopravvivere. Altri individui non sono adatti e perciò muoiono, oppure non trovano un modo di continuare la discendenza. Se la riproduzione avesse l'unico pregio di conservare i tratti migliori di una specie, a seguito di un cambiamento ambientale, ormai "perfezionatasi" a vivere in un dato ambiente, la specie andrebbe incontro all'estinzione. Accade in realtà che nel patrimonio genetico si generino delle mutazioni casuali, che, aumentando la variabilità all'interno della popolazione, danno vita a nuovi organismi in potenza adatti o meno ad eventuali modificazioni ambientali.

Allo stesso modo, in un algoritmo genetico, viene generata una "popolazione" di $N$ possibili soluzioni al problema, ognua delle quali è più o meno "adatta" a risolverlo. Nel caso del commesso viaggiatore, le soluzioni sono i percorsi, ossia elenchi di punti visitati in un dato ordine e il "livello di adattamento" è quantificato dalla lunghezza di ogni percorso. Naturalmente una lunghezza minore corrisponde ad un migliore adattamento.

L'algoritmo ora si sviluppa in diversi step, in ognuno dei quali gli individui si "riproducono" generando due "figli", fino a quando non è stata generata una nuova popolazione di $N$ elementi. Gli individui più adatti devono avere maggiore probabilità di riprodursi. Sono stati implementati due diversi algoritmi di selezione:
1. La probabilità è proporzionale al reciproco della lunghezza del percorso. In questo modo quelli con lunghezza minore verranno scelti più spesso;
2. non viene associata una probabilità di essere scelti per la riproduzione. L'intera popolazione viene suddivisa in due sottogruppi casuali, con dimensione compresa tra il 5% e il 10% di quella totale. Da entrambi viene scelto il percorso migliore per la riproduzione.

La riproduzione consiste di due fasi: crossing-over e mutazione. Il primo imita il crossing-over naturale, durante il quale cromosomi omologhi scambiano una porzione del propio materiale genetico. In natura le porzioni scambiate appartengono allo stesso gene, perciò l'effetto è una ricombinazione degli alleli. Si tratta di un vincolo che mantiene coerente la struttura del cromosoma. Anche nell'algoritmo genetico questo scambio deve essere soggetto ad un vincolo, che nel nostro caso è la necessità di visitare ogni punto una e una sola volta, ossia i punti non devono duplicarsi, né eliminarsi. Inoltre aggiungiamo la restrizione che il primo punto visitato sia sempre lo stesso. Il corssing-over altera ciascuno dei due percorsi (dei "genitori") a partire da un punto casuale: i successivi vengono mescolati in modo che assumano lo stesso ordine presente nell'altro individuo.

Terminato il corssing-over i nuovi individui vengono fatti mutare, secondo 5 modalità diverse, che possono anche concorrere:
* Scambio: vengono scambiati due punti a caso all'interno del percorso;
* scorrimento: a partire da un punto casuale, vengono spostati $M$ ($M < N - 1$) punti consecutivi di $L$ posizioni ($L < N - 1$). Lo scorrimento è ciclico;
* permutazione: vengono scelte due sotto-sequenze di $M$ punti non sovrapposti ($M < N/2 - 1$) e scambiati tra di loro;
* inversione: viene invertito l'ordine di $M$ punti consecutivi ($M < N - 1$).

La probabilità che si verifichi ogni mutazione è rispettivamente.

Poniamo $N = 300$ è il numero di punti del percorso. Eseguiamo l'algoritmo per $200$ passi e in ognuno salviamo la lunghezza del percorso migliore. Infine mostreremo due grafici che rappresentano il migliore percorso casuale generato all'inizio e quello migliore trovato alla fine dell'algoritmo.

### Caso 1: i punti giacciono su una circonferenza

Dopo una serie di tentativi, le probabilità per ogni mutazione sono state poste come segue:
* Scambio: 1.5%;
* scorrimento: 0.1%;
* permutazione: 1%;
* inversione: 1.5%.

Il crossingover non viene eseguito ogni volta che due cromosomi sono stati selezionati, ma con una probabilità del $70%$.

La stessa popolazione viene ottimizzata usando i due diversi metodi di selezione proposti, a parità degli altri parametri (dimensione della popolazione, numero di iterazioni, probabilità che venga eseguito il crossingover, probabilità di mutazione)

```python
import matplotlib
import matplotlib.pyplot as plt
import numpy as np
```

```python
step, fit = np.loadtxt("./best_fit_outOLD.dat", usecols=(0,1), delimiter = ",",unpack='true')
step, fitav = np.loadtxt("./best_av_fit_outOLD.dat", usecols=(0,1), delimiter = ",",unpack='true')

step, fitt = np.loadtxt("./best_fit_outt.dat", usecols=(0,1), delimiter = ",",unpack='true')
step, fitavt = np.loadtxt("./best_av_fit_outt.dat", usecols=(0,1), delimiter = ",",unpack='true')

fig, (plot, pltt) = plt.subplots(1,2)
fig.set_figwidth(15)
fig.set_figheight(7)
plot.set_title('Metodo di selezione (1)')                                                                        
plot.plot(step,fit, label = "Lunghezza migliore per step")
plot.plot(step,fitav, label = "Lunghezza media sulla metà migliore della popolazione")
plot.legend(loc = 'best')
plot.set_xlabel("# step")
plot.set_ylabel("Distanza migliore nella popolazione")
pltt.set_title('Metodo di selezione (2)')                                                                        
pltt.plot(step,fitt, label = "Lunghezza migliore per step")
pltt.plot(step,fitavt, label = "Lunghezza media sulla metà migliore della popolazione")
pltt.legend(loc = 'best')
pltt.set_xlabel("# step")
pltt.set_ylabel("Distanza migliore nella popolazione")

print("Metodo 1:")
print(" Lunghezza del percorso proposto come soluzione: " + str(fit[len(fit)-1]))
print(" Lunghezza media della metà dei percorsi migliori: " + str(fitav[len(fitav)-1]))
print("Metodo 2:")
print(" Lunghezza del percorso proposto come soluzione: " + str(fitt[len(fitt)-1]))
print(" Lunghezza media della metà dei percorsi migliori: " + str(fitavt[len(fitavt)-1]))
```

Chiaramente il secondo metodo di selezione converge più rapidamente ad un _plateau_, che non è detto sia la soluzione al problema. Solamente nel caso di punti su una circonferenza, è possibile verificare se la soluzione proposta sia quella effettiva. Il percorso migliore infatti dovrebbe passare per punti consecutivi ordinatamente. Visualizziamo sul piano cartesiano i punti generati e il percorso che li collega.

```python
x, y = np.loadtxt("./BestOLD.dat", usecols=(0,1), delimiter = ",",unpack='true')
x = np.append(x, x[0]); y = np.append(y, y[0])
xp, yp = np.loadtxt("./Best_preOLD.dat", usecols=(0,1), delimiter = ",", unpack ='true')
xp = np.append(xp, xp[0]); yp = np.append(yp, yp[0])

xt, yt = np.loadtxt("./Bestt.dat", usecols=(0,1), delimiter = ",",unpack='true')
xt = np.append(xt, xt[0]); yt = np.append(yt, yt[0])
xpt, ypt = np.loadtxt("./Best_preOLD.dat", usecols=(0,1), delimiter = ",", unpack ='true')
xpt = np.append(xpt, xpt[0]); ypt = np.append(ypt, ypt[0])

fig, ((plt_pre, plt_post),(plt_pret, plt_postt)) = plt.subplots(2,2)
fig.set_figwidth(20)
fig.set_figheight(20)

plt_post.set_title("Percorso migliore al termine dell'algoritmo - metodo (1)")
plt_post.plot(x,y, 'go-')
plt_post.set_xlabel("x")
plt_post.set_ylabel("y")
plt_pre.set_title("Percorso migliore prima dell'esecuzione dell'algoritmo - metodo (1)")
plt_pre.plot(xp,yp, 'ro-')
plt_pre.set_xlabel("x")
plt_pre.set_ylabel("y")

plt_postt.set_title("Percorso migliore al termine dell'algoritmo - metodo (2)")
plt_postt.plot(xt,yt, 'go-')
plt_postt.set_xlabel("x")
plt_postt.set_ylabel("y")
plt_pret.set_title("Percorso migliore prima dell'esecuzione dell'algoritmo - metodo (2)")
plt_pret.plot(xpt,ypt, 'ro-')
plt_pret.set_xlabel("x")
plt_pret.set_ylabel("y")
```

Non sono visibili sovrapposizioni di linee nel secondo grafico e i punti sembrano essere stati raggiunti tutti consecutivamente. Il programma ha trovato, se non l'esatta soluzione al problema, una molto vicina.

Il primo metodo di selezione è più veloce del secondo, ma necessita di più passi e di una popolazione più numerosa affinché restituisca un tentativo di soluzione. Impostiamo $N$ a 1500 individui e $1000$ passi per l'algoritmo. Confrontiamo quindi i nuovi risultati coi precedenti.

Come tutti gli algoritmi di ottimizzazione che lavorano in maniera approssimata, ossia senza provare tutte le risposte possibili (dove ce ne sia effetivamente un numero massimo finito), gli algoritmi genetici non sono nati per la ricerca della soluzione assoluta. Riconosciuto che per alcuni problemi una ricerca esatta non è tentabile, perché richiederebbe tempi su scale superiori a quelle della vita umana, gli algoritmi approssimati possono dare una soluzione ragionevole, vicina a quella esatta. 

Per quanto riguarda in particolare gli algoritmi genetici, c'è molta libertà interpretativa su come implementare una "selezione", una "riproduzione" o una "mutazione". Sono molte le pubblicazioni che confrontano anche diverse tipologie di crossing-over e generalmente è il problema a determinare quale metodo è più adatto. Nel caso del commesso viaggiatore, è stata proposta una variante alternativa di selezione, che si è dimostrata più efficace di quella generalmente usata per risolvere il problema del commesso viaggiatore.

## Esercizio 2

Il problema di una simulazione che fa uso dell'algoritmo di Metropolis è la correlazione tra stati consecutivi del sistema. Per stimare le grandezze fisiche di nostro interesse, preferiamo lavorare su dati non correlati. Un sistema utile ad aggirare questa correlazione è la divisione in blocchi della simulazione. I valori delle grandezze fisiche istantanee di ogni passo vengono mediate sull'intero blocco. I valori che così si ottengono, se la dimensione del blocco è stata scelta in maniera opportuna, dovrebero essere non correlati e meno dispersi. Vogliamo trovare la lunghezza ottimale dei blocchi e a tal fine calcoliamo le funzioni di autocorrelazione del potenziale.

```python
npart = 108
```

Per tutte e tre la fasi, il sistema viene preparato inizialmente in una configurazione cristallina, con le molecole tra di loro equispaziate. Di seguito una rappresentazione

```python
from mpl_toolkits import mplot3d
%matplotlib inline

fig3D = plt.figure()
ax3D = plt.axes(projection='3d')
fig3D.set_figheight(10)

x, y, z = np.loadtxt("./config.in", usecols=(0,1,2), unpack='true')
ax3D.scatter(x, y, z, marker = "o", s = 100);
ax3D.set_xlabel("x"); ax3D.set_ylabel("y"); ax3D.set_zlabel("z")
plt.title("Configurazione iniziale");
```

Eseguiamo una simulazione di $500000$ passi, in modo da avere un numero sufficientemente grande di dati su cui calcolare la funzione di autocorrelazione. Qui sotto mostriamo un grafio della variazione dell'energia potenziale per particella.

```python
step, tmp_meansol = np.loadtxt("./output/output_prepot_solid.dat", usecols=(0,1), unpack='true')
step, tmp_meanliq = np.loadtxt("./output/output_prepot_liquid.dat", usecols=(0,1), unpack='true')
step, tmp_meangas = np.loadtxt("./output/output_prepot_gas.dat", usecols=(0,1), unpack='true')

T_tar = 0.8

tmp_meansolN = tmp_meansol/npart
tmp_meanliqN = tmp_meanliq/npart
tmp_meangasN = tmp_meangas/npart

plt.figure(figsize=(17,5))
plt.plot(step, tmp_meansolN, label = "Fase solida");
plt.plot(step, tmp_meanliqN, label = "Fase liquida");
plt.plot(step, tmp_meangasN, label = "Fase gassosa");
# plt.plot(step, T_tar*np.ones(step.size), color = 'green', label = "Target temperature");
plt.title("Energia Potenziale");
plt.xlabel("Simulation step")
plt.ylabel("V [reduced units]")
plt.legend();
plt.grid();
```

La funzione da studiare è la seguente:

$$ \chi(t) = \frac{<U(t')U(t' + t)>_{t'} - <U> }{\sigma^2_U} $$

Siccome per ogni valore di $t$ deve essere calcolato il valor medio su $t'$, il tempo richiesto al calcolo di questa funzione, per ogni $t \in {0,1,\dots, 5\cdot 10^5}$, sarebbe eccessivo. Ci limitiamo allora ai primi $500$ valori, con la speranza che l'autocorrelazione tra i dati decresca rapidamente prima di raggiungere questo limite.

```python
mean = 0
mean2 = 0

tmax = 500

datas = [tmp_meansolN, tmp_meanliqN, tmp_meangasN]
mean = np.zeros(3)
mean2 = np.zeros(3)

for k in range(len(datas)):
    for t in range(len(step)):
        # Somma parziale per la media di U al variare di t
        mean[k] += datas[k][int(t)]
        mean2[k] += datas[k][int(t)]**2
    mean[k] /= len(step)
    mean2[k] /= len(step)

mean_prod = [np.zeros(len(step)) for i in range(len(datas))]

for k in range(len(datas)):
    for t in range(tmax):
        # Valor medio di U(t')*U(t' + t) al variare di t'
        times_ = np.arange(0, len(step) - t, 1)
        for t_ in times_:
            mean_prod[k][int(t)] += datas[k][int(t_)]*datas[k][int(t_ + t)]
        mean_prod[k][int(t)] /= len(times_)
        print(str(t) + ", data " + str(k), end = ' \r')
```

```python
# Calcolo della varianza
sigma = mean2 - np.power(mean, 2)
X = [0,0,0]
for i in range(len(datas)):
    X[i] = (mean_prod[i] - mean[i]**2)/sigma[i]

# Salvo i risultati ottenuti
np.savetxt('CHI.dat', X)
```

```python
plt.figure(figsize = (17, 5) )
for i in range(len(datas)):
    plt.plot(step[:tmax], X[i][:tmax] )
plt.title('Autocorrelazione')
plt.xlabel('$t$')
plt.ylabel("$\chi(t)$");
```

È evidente che già dopo 100 passi la correlazione è praticamente assente nel caso del solido. Ce ne vogliono circa $300$ affinché anche per il liquido si possa trascurare. Nel caso del gas, l'analisi per per scorrimenti "temporali" inferiori a $500$ passi di simulazione non è sufficiente a stabilire quando l'autocorrelazione diventa (quasi) nulla. Dato che per $t$ superiori a $500$ il calcolo impiega troppo tempo, chiediamo al lettore di attendere fino alla prossima analisi per giungere a delle considerazioni anche per il caso del gas. Per lo stato solido e quello liquido, blocchi di dimensione $100$ passi e $300$ passi sono allora sufficienti a produrre dati scorrelati.

Osserviamo ora come varia l'incertezza sulla stima finale dell'energia potenziale al variare del numero di blocchi (o equivalentemente al variare della loro dimensione). Se $N$ è il numero di blocchi e $M$ il numero di passi totali della simulazione (qui $N = 500000$), $L = M/N$ è la lunghezza di ogni blocco. Per diversi valori di $L$ compresi tra $L=10$ e $L=10000$, produciamo un grafico della variazione dell'incertezza su $<U>$.

```python
M = len(step)
# lenght = np.array([10, 50, 100, 200, 400, 500, 800, 1000])
lenght = np.array([i for i in range(10, 10001) if not(500000%i) ])
nblk = M/lenght
print(nblk)
sigma2 = [np.zeros(len(lenght)) for i in range(len(datas))]
sigma = [np.zeros(len(lenght)) for i in range(len(datas))]
for k in range(len(datas)):
    for i_L in range(len(lenght)):
        mean = 0
        mean2 = 0
        for iblk in range(int(nblk[i_L])): # iblk è il numero del blocco attuale
            average = 0
            for i in range(lenght[i_L]): # i scorre tra 0 e la lunghezza del blocco
                average += datas[k][iblk*lenght[i_L] + i]
            average /= lenght[i_L]
        # for inlk in range(lenght[i_L]):
            mean += average
            mean2 += average**2
        sigma2[k][i_L] = mean2/nblk[i_L] - (mean/nblk[i_L])**2
        sigma[k][i_L]= np.sqrt(sigma2[k][i_L]/(nblk[i_L]-1))
        
```

```python
plt.plot(lenght, np.transpose(sigma), 'o-')
plt.title("Variazione della deviazione standard rispetto alla dimensione del blocco")
plt.xlabel("Dmensione del blocco")
plt.ylabel("Dev.std.")
```

L'incertezza aumenta per dimensioni del blocco piccole, ma tende ad un _plaeau_ per dimensioni grandi. Il calcolo dell'incertezza infatti non ha tenuto conto della correlazione tra i dati, che è presente per blocchi piccoli. Poiché l'autocorrelazione è positiva, per $L$ piccoli otteniamo una sottostima dell'incertezza. Superato un certo valore soglia (che per la fase solida e liquida abbiamo osservato essere circa $100$ e $300$ passi), l'autocorrelazione diventa praticamente assente, così la stima cell'incertezza diventa più precisa e si assesta intorno ad un valore fisso.

Dal grafico deduciamo anche la dimensione ideale dei blocchi nel caso del gas - $L $~$6000$ - che non era stato individuato nell'analisi precedente sull'autocorrelazione.

## Esercizio 4

Infine eseguiamo una simulazione di tre fasi dell'Argon ($\sigma = 0.34$ nm, $\epsilon/k_B = 120$ K, $m=39.948$ amu), nelle condizioni seguenti:
1. fase solida: $\rho^\star = 1.1$, $T^\star = 0.8$ (raggio di cut-off: $r_c = 2.2$)
2. fase liquida: $\rho^\star = 0.8$, $T^\star = 1.1$ (raggio di cut-off: $r_c = 2.5$)
3. fase gassosa: $\rho^\star = 0.05$, $T^\star = 1.2$ (raggio di cut-off: $r_c = 5.0$)

Misuriamo l'energia potenziale per particella e la pressione del gas, sia con l'algoritmo di Metropolis che con il programma di dinamica molecolare (dall'esercitzione 4). Includiamo inoltre il calcolo della funzione di distribuzione radiale.

Ci riferiamo sempre all'esercitazione 4 per i valori di temperatura da cui partire per ottenere degli stati di equilibrio con la temperatura di lavoro corretta nel caso della simulazione con la dinamica molecolare.

La funzione di distribuzione radiale permette di stimare le correzioni da applicare ai valori calcolati di pressione e energia potenziale , tenendo condo dei contributi delle code di potenziale esterne al raggio di cut-off. Se $r_c$ è il potenziale di cut-off, la correzioni per l'energia potenziale (per particella)  è la seguente:

$$ \frac{<V>_{\text{tail}}}{N} = 2\pi\rho N\int_{0}^{\infty}dr r^2 V(r)g(r).$$

Siccome le code si riferiscono a grandi distanze, la funzione di distribuzione radiale tende a quella del gas perfetto $g(r)\approx 1$, perciò la correzione diventa:

$$ \frac{<V>_{\text{tail}}}{N} \approx 2\pi\rho N\int_{0}^{\infty}dr r^2 V(r) = \frac{8\pi\rho\epsilon}{3r_c^3}\big( \frac{1}{3r_c^6} - 1 \big) $$

Per la pressione la correzione si trova usando il teorema del viriale:

$$ \frac{<w>_{\text{tail}}}{3N\epsilon} \approx \frac{2}{3}\pi\rho\int_{r_c}^{\infty}dr r^2 48\big[ \big(\frac{\sigma}{r}\big)^{12} - 0.5\big(\frac{\sigma}{r}\big)^6 \big] = 32\pi \rho \big[ \frac{1}{9r_c^9} - \frac{1}{6r_c^3} \big]$$

### Fase solida

## Conclusioni

Le stime della funzione di distribuzione radiale ottenute con Metropolis e con il programma di diamica molecolare sono tra loro compatibili. Notiamo inoltre che in tutte e tre le simulazioni, la distribuzione di moleole tra loro distanti meno di un certo valore radiale è nulla, effetto della repulsione a $r$ piccoli.

Le funzioni di distribuzione radiale della fase solida e della liquida presentano più picchi, che riflettono la struttura ordinata delle molecole. Tutte le funzioni per $r$ grandi tendono al caso del gas ideale, in cui le molecole sono distribuite uniformemente per ogni $r$.
